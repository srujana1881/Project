# Feature Engineering & Preprocessing

```python
# Feature Vector
feature_cols = df.columns[1:]

assembler = VectorAssembler(
    inputCols=feature_cols,
    outputCol="features_raw"
)
```

```python
# Scaling
scaler = StandardScaler(
    inputCol="features_raw",
    outputCol="features",
    withStd=True,
    withMean=False
)
```

```python
pipeline = Pipeline(stages=[assembler, scaler])
```

```python
pipeline_model = pipeline.fit(df)
processed_df = pipeline_model.transform(df)
```

```python
processed_df.select("label","features").show(5)
```

```python
processed_df.persist()
```

```python
quality_df = processed_df.describe()
quality_df.toPandas().to_csv("data_quality.csv", index=False)
```

```python
class_dist = processed_df.groupBy("label").count()
class_dist.toPandas().to_csv("class_distribution.csv", index=False)
```
